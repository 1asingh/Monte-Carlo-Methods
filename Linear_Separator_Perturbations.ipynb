```python
# Generating Data Points for SVM classifier which gives the line y = x
samples = 150
x_plus = np.random.uniform(-1, 1, samples)
x_minus = np.random.uniform(-1, 1, samples)

y_plus = x_plus + 1
y_minus = x_minus - 1
x = np.zeros((2*samples, 2))

labels = np.zeros(2*samples)
print(labels)
for i in range(2*samples):
  if i%2 == 0:
    x[i, 0] = x_plus[i//2]
    x[i, 1] = y_plus[i//2]
    labels[i] = 1
  if i%2 == 1:
    x[i, 0] = x_minus[i//2]
    x[i, 1] = y_minus[i//2]
    labels[i] = -1
x_new = np.copy(x)
print(labels)
plt.scatter(x[:, 0], x[:, 1])
plt.show()
```

```python
from sklearn.svm import SVC
SVM = SVC(kernel='linear')
SVM.fit(x, labels)
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
plt.scatter(x[:, 0], x[:, 1])
plot_svc_decision_function(SVM)
print(x_new)

```

```python
# Simulate gaussian noise with mean 0 and variance 1
sampling_count = 1000
distribution = np.array([])
for j in range(sampling_count):
    gauss_noise = np.random.normal(0, 1, 2*samples)
    for i in range(2*samples):
        x_new[i, 1] = x[i,1] + gauss_noise[i]
    SVM = SVC(kernel='linear')
    SVM.fit(x_new, labels)
    plt.scatter(x_new[:, 0], x_new[:, 1])
    plot_svc_decision_function(SVM)
    distribution = np.append(distribution, -1/SVM.coef_[0][0])
    print("Simulations run: ", j)
```
